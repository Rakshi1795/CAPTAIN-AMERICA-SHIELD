```python
import turtle as t

t.bgcolor=('black')

def circle(radius,color):
    t.color(color)
    t.begin_fill()
    t.circle(radius)
    t.end_fill()
    
t.goto(-30,-280)
circle(300,'red')

t.goto(-30,-240)
circle(260,'white')

t.goto(-30,-200)
circle(220,'red')

t.goto(-30,-160)
circle(180,'blue')

t.goto(-200,75)
t.begin_fill()
t.color('white')

for i in range(5):
    t.forward(340)
    t.right(144)
t.end_fill()

t.hideturtle()
t.done()
```
